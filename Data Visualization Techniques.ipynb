```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("sorted_data.csv")

# Keep numeric data
num = df.select_dtypes(include='number')

# -----------------------------
# 1. Line Plot
# -----------------------------
plt.plot(num.iloc[:,0])
plt.title("Line Plot")
plt.show()


# 2. Bar Plot

num.iloc[:,0].value_counts().head().plot(kind='bar')
plt.title("Bar Plot")
plt.show()


# 3. Histogram

plt.hist(num.iloc[:,0])
plt.title("Histogram")
plt.show()


# 4. Scatter Plot

plt.scatter(num.iloc[:,0], num.iloc[:,1])
plt.title("Scatter Plot")
plt.show()
```
